```python

import pandas as pd
#Link de repositorio: carpeta
link_folder= 'https://github.com/personafujoshi/prueba_pc3/tree/main'
#Link de repositorio: archivos
link_file1=pd.read_excel('https://github.com/personafujoshi/prueba_pc3/blob/main/level1.xlsx?raw=true',header=None)
link_file2=pd.read_excel('https://github.com/personafujoshi/prueba_pc3/blob/main/level2.xlsx?raw=true',header=None)
link_file3=pd.read_excel('https://github.com/personafujoshi/prueba_pc3/blob/main/level3.xlsx?raw=true',header=None)
link_file4=pd.read_excel('https://github.com/personafujoshi/prueba_pc3/blob/main/level4.xlsx?raw=true',header=None)
link_file1
```

```python
#LIMPIANDO "LEVEL 1"
new_title = link_file1.iloc[4]
link_file1.columns=new_title
link_file1.reset_index(drop=True, inplace=True)
n1_file1
n2_file1 = n1_file1[~(n1_file1.iloc[:, 2:].isna().all(axis=1) & n1_file1.iloc[:, 1].notna())]
n2_file1
n2_file1 = n2_file1.loc[:, n2_file1.iloc[0].notna()]
n2_file1.reset_index(drop=True, inplace=True)
n2_file1
n2_file1 = n2_file1.drop(index=[0, 1, 2])
n2_file1.reset_index(drop=True, inplace=True)
n2_file1 = n2_file1.iloc[:, :-1]
n2_file1
```

```python
n2_file1.info()
```

```python
for col in n2_file1.columns[2:8]:
    n2_file1[col] = pd.to_numeric(n2_file1[col], errors='coerce')

n2_file1
```

```python
n2_file1.info()
```

```python
#LIMPIANDO "LEVEL 2"

encabezado=link_file2.iloc[0]
link_file2.columns=encabezado

link_file2= link_file2.drop(index=[0, 1, 2])
link_file2.reset_index(drop=True, inplace=True)

n1_file2= link_file2[~(link_file2.iloc[:, 1:].isna().all(axis=1) & link_file2.iloc[:, 0].notna())]
n1_file2 = n1_file2.loc[:, n1_file2.iloc[0].notna()]
n1_file2 = n1_file2.drop(columns=link_file2.columns[6])
n1_file2 = n1_file2.iloc[:, :-1]

for col in n1_file2.columns[2:8]:
    n1_file2[col] = pd.to_numeric(n1_file2[col], errors='coerce')
n1_file2.head()
n1_file2.info()
n1_file2
```

```python
#LIMPIANDO LEVEL 3
link_file3
new_tittle3 = link_file3.iloc[0]
link_file3.columns=new_tittle3
n1_file3=link_file3.drop(index=[0,1,2])
link_file3.reset_index(drop=True, inplace=True)
n1_file3
n2_file3 = n1_file3[~(n1_file3.iloc[:, 2:].isna().all(axis=1) & n1_file3.iloc[:, 1].notna())]
n2_file3
n2_file3 = n2_file3.loc[:, n2_file3.iloc[0].notna()]
n2_file3.reset_index(drop=True, inplace=True)
n2_file3
n2_file3 = n2_file3.iloc[:, :-1]
n2_file3
n2_file3.info()
```

```python
for col in n2_file3.columns[2:8]:
    n2_file3[col] = pd.to_numeric(n2_file3[col], errors='coerce')

n2_file3
```

```python
n2_file3.info()
```

```python
#LIMPIANDO LEVEL 4
encabezado=link_file4.iloc[0]
link_file4.columns=encabezado
link_file4 = link_file4.drop(index=0)
link_file4.reset_index(drop=True, inplace=True)
n2_file4= link_file4[~(link_file4.iloc[:, 2:].isna().all(axis=1) & link_file4.iloc[:, 1].notna())]
n2_file4 = n2_file4.loc[:, n2_file4.iloc[0].notna()]
n2_file4.reset_index(drop=True, inplace=True)
column_name = n2_file4.columns[2]
n2_file4 = n2_file4.drop(columns=[column_name])
for col in n2_file4.columns[2:8]:
    n2_file4[col] = pd.to_numeric(n2_file4[col], errors='coerce')

n2_file4
```

```python
n2_file4.info()
```

```python
#INTEGRANDO
#en lista

joins=[n2_file1, n1_file2, n2_file3, n2_file4]
allColumnNames=[]
for df in joins:
    allColumnNames.append(set(df.columns))

# verificando si tienen la misma cantidad
[len(cols) for cols in allColumnNames]
```

```python
#concatenar
conc_joins = pd.concat(joins, ignore_index=True)
conc_joins
```

```python
conc_joins.info()
```

```python
for col in conc_joins.columns[2:8]:
    conc_joins[col] = pd.to_numeric(conc_joins[col], errors='coerce')

conc_joins
```

```python
conc_joins.info()
```

```python
conc_joins['Country'] = conc_joins['Country'].str.replace(r'\([^)]*\)', '', regex=True)
conc_joins['Country'] = conc_joins['Country'].str.split(',').str[0]
conc_joins['Country'] = conc_joins['Country'].str.upper()
conc_joins
```

```python
#MERGE ENTRE "CONC_JOINS" Y ARCHIVO "COUNTRYCODES"
xyz2nd= pd.read_excel('https://github.com/personafujoshi/prueba_pc3/blob/main/countryCODES.xlsx?raw=true', header=None)
xyz2nd
title_xyz2nd=xyz2nd.iloc[0]
xyz2nd.columns=title_xyz2nd
xyz2nd

#mergemerge
merge=conc_joins.merge(xyz2nd,left_on='Country',right_on='Country')
merge
```
